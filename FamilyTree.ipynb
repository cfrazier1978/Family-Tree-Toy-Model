# Family Tree Toy Model
## Goal
The purpose of this model is to play out a game of families coming together and taking on a certain last name.   As you add in more families, which family name will last until the end of the game?

Many elements of this model will be random.  The gender of children, the child’s life preferences and who they are matched with will all be randomly assigned.

This model could be adapted to track any sort of change of state of a population or culture over time.
Background
Also known as a surname, the family name is the portion of a persons ‘personal name’ that indicates your family, tribe or community.  A surname indicates genetic inheritance, identifying members of a family unit due to marriage and changing one’s surname, or by birth and being given your surname by your parents.

Different cultures have their own traditions on how names are constructed.  The family name could appear at the start, middle or end of the person’s name.  Sometimes a person has several surnames.  This could be by choice due to multiple relationships or because of their culture.  In traditional Spanish culture, multiple family surnames are hyphenated together.

## Model Rules
The game will begin with a set number of couples, each with a different last name.  Each timestep will be one generation of the family tree.  During each generation, the family will have a random number of children who will have a random gender assigned.  They will then seek out a spouse from a different family.

For the sake of simplicity of the initial model construction, we will keep everything very traditional.  A new family is created by joining a male and a female from different families.  The family will be assigned the last name of the male family member.  These new families will then have children who will seek out their own spouses to make a family.

Once the model works, other social factors can be added where same sex couples could be joined together and either have their own children or adopt.  The method for having children isn’t important, only that they have children who then take on their family name.  I will also add in the option that a family chooses to have zero children.

Someone from a family can not join with another person with the same last name, even if, generationally, that would be acceptable.
## Assumptions
N/A, all the variables of this model are known.
Random Variables
Number of children: This will be a random number with a ceiling.  For the first few experiments, there will also be a set minimum so that there isn't an arbitrary ending of the model because a family has zero children.
Gender of child: Randomly assign if it is a boy or girl.
Marriage Preference: In future iterations of the model, children will have a choice on whether they want to create a new family or not.
Partner Preference: In future iterations of the model, a gender preference for a spouse will be added to add to the possibilities of outcomes to the model.
## Initial Conditions
Number of families: Each experiment will have a set number of starting families
Family last name: A name assigned to each starting family
Could be as simple as A and B for simplicity sake
Number of Children Range: The range for the random variable to set a number of children per family
Number of timesteps: Timesteps will be defined as generations.
Marriage Preference: This will be a percent probability that a child will want to get married and create a family.
Partner Preference: This will be a percent probability of the sex of the partner that the child prefers to start a family with.
## Model Considerations
I need to figure out how I can put all the eligible bachelors / bachelorettes into a list and randomly assign relationships.  This will be further complicated when I allow same sex relationships.

If a same sex couple is joined, need to decide how to assign a last name.  It could be random between their last names, or they could be assigned a new family name.  The choice to create their own family name should be an option given to non-same sex couples as well.
## Variables to Track
Number of Families per Family Name: Track the population of a family name during each generation (timestep)
Children Without Spouses: How many children do not get matched, either by choice or because there are no eligible children to be matched with.
